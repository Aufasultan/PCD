```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import block_reduce


img = cv2.imread(cv2.samples.findFile("{✨°pinterest_@keriaah°✨}.jpeg"))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()


# Down Sampling

down_max = block_reduce(img, block_size=(2,2,1), func=np.max)

# Average Pooling
down_avg = block_reduce(img, block_size=(2,2,1), func=np.mean).astype(np.uint8)

# Median Pooling
down_med = block_reduce(img, block_size=(2,2,1), func=np.median).astype(np.uint8)

fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(down_max); ax[0].set_title("Down Sampling - Max")
ax[1].imshow(down_avg); ax[1].set_title("Down Sampling - Average")
ax[2].imshow(down_med); ax[2].set_title("Down Sampling - Median")
for a in ax: a.axis("off")
plt.show()


# Up Sampling

# Nearest Neighbor
up_nn = cv2.resize(down_avg, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)

# Bilinear
up_bilinear = cv2.resize(down_avg, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_LINEAR)

# Bicubic
up_bicubic = cv2.resize(down_avg, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_CUBIC)

fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(up_nn); ax[0].set_title("Up Sampling - Nearest Neighbor")
ax[1].imshow(up_bilinear); ax[1].set_title("Up Sampling - Bilinear")
ax[2].imshow(up_bicubic); ax[2].set_title("Up Sampling - Bicubic")
for a in ax: a.axis("off")
plt.show()
```
